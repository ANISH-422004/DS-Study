```python
from matplotlib import text


class NLPApp:
    def __init__(self):
        self.__database = {}
        self.__first_menu()

    def __first_menu(self):
        choice = input(
            " Hi there! Welcome to NLP App. How would you like to proceed?  1.Not a member ? Sign up 2.Existing Member ? Login  3.Exit  "
        )

        if choice == "1":
            self.__register()
        elif choice == "2":
            self.__login()
        else:
            self.exit()

    def __second_menu(self):
        choice = input(
            " Choose an option: 1. NER 2. Language Detection 3. Sentiment Analysis 4. Exit "
        )

        if choice == "1":
            self.__ner()
        elif choice == "2":
            self.__language_detection()
        elif choice == "3":
            self.__sentiment_analysis()
        else:
            self.exit()

    def __register(self):
        name = input("Enter your name: ")
        email = input("Enter your email: ")
        password = input("Create a password: ")

        if email in self.__database:
            print("Email already registered. Please login.")
            self.__login()
            return

        self.__database[email] = {"name": name, "password": password}
        self.__first_menu()

    def __login(self):
        email = input("Enter Your email")
        if email in self.__database:
            password = input("Enter your password: ")
            if password == self.__database[email]["password"]:
                print(f"Welcome back, {self.__database[email]['name']}!")
                self.__second_menu()
            else:
                print("Incorrect password. Please try again.")
                self.__login()

    def __ner(self):
        text = input("Enter paragraph: ")
        target_word = input("Enter the target word which whould you like to search : ")
        client = nlpcloud.Client(
            "llama-3-1-405b", "cbb5d8b65ba7003d02402a323bdbd467bbdf7c29", gpu=True
        )
        ans = client.entities(text, searched_entity=target_word)
        print(ans)

    def __sentiment_analysis(self):
        text = input("Enter paragraph: ")
        client = nlpcloud.Client(
            "gpt-oss-120b", "cbb5d8b65ba7003d02402a323bdbd467bbdf7c29", gpu=True
        )
        ans = client.sentiment(
            text,
            target="NLP Cloud",
        )

        print(ans)

    def exit(self):
        print("Thank you for using NLP App. Goodbye!")


obj = NLPApp()
```
